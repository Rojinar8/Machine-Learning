## **Group by and Aggregation**

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Flavors.csv')
df
```

```python
group_by_frame = df.groupby('Base Flavor')
group_by_frame.sum()
```

```python
group_by_frame.mean("/content/drive/MyDrive/Flavors.csv")

```

```python
group_by_frame.mean(numeric_only= True)
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({
    'Flavor Rating':['mean', 'max', 'count', 'sum'],
    'Texture Rating':['mean', 'max', 'count','sum']
})
```

```python
df.groupby('Base Flavor').describe()
```

## **Merge,Join and Concatenate**

```python
import pandas as pd
df1 = pd.read_csv("/content/drive/MyDrive/LOTR.csv")
df1
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/LOTR 2.csv")
df2

```

```python
pd.concat([df1, df2], ignore_index= True)
```

```python
#Inner join
df1.merge(df2, how= 'inner', on = ['FellowshipID','FirstName'])

```

```python
#Outer join
df1.merge(df2, how= 'outer')
```

```python
#Left join
df1.merge(df2, how= 'left')
```

```python
#Right join
df1.merge(df2, how= 'right')

```
