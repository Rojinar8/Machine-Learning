```python
#import dataset customer call list
import pandas as pd
df =pd.read_excel("/content/drive/MyDrive/Customer Call List.xlsx")
df
```

```python
# Drop Unnecessary Column
df.drop(columns=['Not_Useful_Column'], inplace=True, errors='ignore')
df

```

```python
# Fill missing Last_Name with "Unknown"
df['Last_Name'].fillna("Unknown", inplace=True)
df
```

```python
# Drop rows where Phone_Number is missing
df.dropna(subset=['Phone_Number'], inplace=True)
df
```

```python
# Fill missing Do_Not_Contact with "No"
df['Do_Not_Contact'].fillna("No", inplace=True)
df
```

```python
# removing unnecessary character from phone number
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
df
```

```python
# Convert Yes/No to Y/N in Paying Customer
df['Paying Customer'] = df['Paying Customer'].str.strip().replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
# Convert Y/N to Yes/No in Do_Not_Contact
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.strip().replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
# Remove special characters (e.g., /) from Last_Name
df['Last_Name'] = df['Last_Name'].str.replace(r'[^a-zA-Z ]', '', regex=True).str.strip()
df
```

```python
df
```

```python
# Make phone number format as 3digits- 3digits- remaining number
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})-?(\d{3})-?(\d{4}).*', r'\1-\2-\3', regex=True)
df
```

```python
# drop duplicates
df.drop_duplicates(inplace=True)
df
```

```python
# for address fill Nan, N/A, n/a with blank
df['Address'] = df['Address'].replace(['nan', 'N/A', 'n/a', 'N/a'], '')
df['Paying Customer'] = df['Paying Customer'].replace(['nan', 'N/A', 'n/a', 'N/a'], '')
df
```

```python
# remove the entire row where there is no phone number
df = df[df['Phone_Number'] != '']
df

```

```python
# split the column Address and create three columns named: stress_address, state, zip_code
df[['street_address', 'state', 'zipcode']] = df['Address'].str.split(',', expand=True, n=2)
df['zipcode'] = df['zipcode'].str.strip()  # Remove leading/trailing spaces from zipcode
df['state'] = df['state'].str.extract(r'([A-Z]{2})')  # Extract 2-letter state abbreviation using regex
df['street_address'] = df['street_address'].str.strip()  # Remove leading/trailing spaces from street_address
df
```

```python
# remove rows if Do Not Contact has the value 'Y'
df = df[df['Do_Not_Contact'] != 'Y']
df
```
