## Data Cleaning with pandas

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/pandas_dataset.csv")
df.head(5)
```

```python
df["Date"]=df["Date"].replace("'","",regex=True)
df.head(5)
```

```python
df
```

```python
#return a dataset with no empty cells
#df.dropna() will remove empty cells but will not change the dataframe
#If we want dataframe to be changed,use inplace=true
new_df=df.dropna()
new_df
```

```python
# another way of handling the null values is by replacing the null values with some other elements
df_new2=df.fillna(130)
df_new2
```

```python
#fill specific column's null values
df
```

```python
df['Calories'].fillna(150,inplace=True)
df
```

```python
#fill specific null values with mean,median or mode
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/pandas_dataset.csv")
x=df["Calories"].mean()
df["Calories"].fillna(x,inplace=True)
df
```
