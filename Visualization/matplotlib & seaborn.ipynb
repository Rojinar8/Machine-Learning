## **Visualization with Matplotlib**

```python
 import matplotlib.pyplot as plt
 import numpy as np
 import pandas as pd
```

```python
#Basics of Matplotlib
#Line plot
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x =np.linspace(0,10,100)
y =np.sin(x)
plt.plot(x, y, label="Sine wave")
plt.title("Line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C','D']
values = [10, 20, 30, 40,]
color = ['olive', 'red', 'black', 'blue']
plt.bar(categories, values, color = color)
plt.title("Bar Chart Example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

## **Visualization with Seaborn**

```python
import seaborn as sns
```

```python
#Histogram and KDE plot:
data=np.random.randn(1000)
sns.histplot(data,kde=True, color='orange')
plt.title("Histogram with KDE")
plt.show()
```

```python
#Heatmap
matrix= np.random.rand(5,5)
matrix
```

```python
sns.heatmap(matrix,annot=True,cmap="magma")
plt.title("Heatmap Example")
plt.show()
```

```python
#pairplot example
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species",markers=["o","s","D"])
plt.show()
```

```python
# read ice cream rating.csv
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Ice Cream Ratings.csv")
df
```

```python
# set date as index
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
print(df)
```

```python
# scatter plot using texture rating and overall rating
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Texture Rating', y='Overall Rating', data=df, color='violet', alpha=0.8)
plt.title('Scatter Plot of Texture Rating vs. Overall Rating (Seaborn)')
plt.xlabel('Texture Rating')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()
```

```python
# hist plot
plt.figure(figsize=(8, 6))
sns.histplot(df['Overall Rating'], kde=True, color='red')
plt.title('Distribution of Overall Ratings (Seaborn)')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot
df["Texture Rating"].plot.box()
```

```python
#pie chart
df["Texture Rating"].value_counts().plot.pie()
```
